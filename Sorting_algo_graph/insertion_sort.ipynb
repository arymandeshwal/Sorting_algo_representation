```python
import time
import random
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
fig.show()
```

```python
xs=[]
arr=[]

for i in range(100):
    xs.append(i)
    arr.append(random.randint(1,101))

for i in range(1, len(arr)): 
    key = arr[i] 
    j = i-1
    while j >= 0 and key < arr[j] : 
        arr[j + 1] = arr[j] 
        j -= 1
    arr[j + 1] = key 
            
    ax.clear()
    ax.bar(xs,arr)
    fig.canvas.draw()
   # time.sleep(0.01)
```

```python
plt.close()
```
