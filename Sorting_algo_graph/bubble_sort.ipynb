```python
import time
import random
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
fig.show()
```

```python
xs=[]
arr=[]

for i in range(100):
    xs.append(i)
    arr.append(random.randint(1,101))


n = len(arr)
for i in range(n):
    for j in range(0, n-i-1):
        if arr[j] > arr[j+1] :
            arr[j], arr[j+1] = arr[j+1], arr[j]
            
    ax.clear()
    ax.bar(xs,arr)
    fig.canvas.draw()
   # time.sleep(0.01)
```

```python
plt.close()
```
