```python
import time
import random
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
fig.show()
```

```python
xs=[]
A=[]

for i in range(100):
    xs.append(i)
    A.append(random.randint(1,101))


for i in range(len(A)): 
    min_idx = i 
    for j in range(i+1, len(A)): 
        if A[min_idx] > A[j]: 
            min_idx = j    
    A[i], A[min_idx] = A[min_idx], A[i] 
            
    ax.clear()
    ax.bar(xs,A)
    fig.canvas.draw()
   # time.sleep(0.01)
```

```python
plt.close()
```
