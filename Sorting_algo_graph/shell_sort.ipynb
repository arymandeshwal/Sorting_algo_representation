```python
import time
import random
```

```python
import matplotlib.pyplot as plt
```

```python
%matplotlib notebook
```

```python
plt.rcParams['animation.html'] = 'jshtml'
```

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
fig.show()
```

```python
xs=[]
arr=[]

for i in range(100):
    xs.append(i)
    arr.append(random.randint(1,101))
 
n = len(arr) 
gap = n//2
while gap > 0:   
    for i in range(gap,n): 
        temp = arr[i] 
        j = i 
        while  j >= gap and arr[j-gap] >temp: 
            arr[j] = arr[j-gap] 
            j -= gap  
        arr[j] = temp 
    gap //= 2
    
    ax.clear()
    ax.bar(xs,arr)
    fig.canvas.draw()
    time.sleep(0.5)
        
        
#ax.clear()
#ax.bar(xs,arr)
#fig.canvas.draw()
#time.sleep(0.01)
```
